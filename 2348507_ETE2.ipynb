## Question 2 -> Custom Activation Function ##

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining Function for standard sigmoid

```python
def sigmoid(x):
    return  1 / (1+np.exp(-x))
```

```python
def sigmoidLeakyRelu(x,alpha=0.01):
    return np.where(x>=0,sigmoid(x),alpha * x + sigmoid(x))
```

Testing with sample input array

```python
input = np.array([-1,0.5,0,0.5,1])
```

Applying with the function

```python
resCustAct = sigmoidLeakyRelu(input)
```

Comparision with the standard sigmoid and relu function

```python
resSigmoid = sigmoid(input)
resRelu = np.maximum(0,input)

print("Displaying the input array:",input)
print("Custom Activation Function: ", resCustAct)
print("Standard sigmoid activation function:", resSigmoid)
print("Standard Relu activation function: ", resRelu)
```

Inference:

The output reflects the characteristics of both the sigmoid and leaky relu functions.
There is not that much difference between standard sigmoid and custom activation function. It provides the combination of smoothness coz of sigmoid and handling neg values effectively by leaky relu. It can learn new patterns.

## Question 3 -> Mini Batch Gradient Descent to minimize the function ##

```python
def func1(x,y):
    return x**2 + y**2
```

```python
def gradientfunc1(x,y):
    return 2*x,2*y
```

Defining the parameters

```python
learning_rate = 0.001
batch_size = 10 
iterations = 100
datapoints = 1000
```

```python
xPoints = np.random.uniform(-10,10,datapoints)
yPoints = np.random.uniform(-10,10,datapoints)
```

Initializing x and y

```python
x = 0
y = 0
```

```python
fValues = []
```

```python
for it in range(iterations):
    
    indices = np.random.permutation(datapoints)
    
    for i in range(0, datapoints, batch_size):
        batch_indices = indices[i:i + batch_size]
        
        gradX, gradY = gradientfunc1(xPoints[batch_indices], yPoints[batch_indices])
        mean_grad_x = np.mean(gradX)
        mean_grad_y = np.mean(gradY)
        
        # Updating the gradients
        
        x -= learning_rate * mean_grad_x
        y -= learning_rate * mean_grad_y

    fValues.append(func1(x,y))
print(f"Final values: x = {x:.4f}, y = {y:.4f}")
```

Inference:

These values indicates that the funnction is converging closer to global minima. I gave the learning rate as 0.001, previously it was 0.05 that gave a huge values.

```python
plt.figure(figsize=(10, 6))
plt.plot(fValues, label='f(x, y) = x^2 + y^2')
plt.xlabel('Iteration')
plt.ylabel('Function value')
plt.legend()
plt.grid(True)
plt.show()
```

Inference:
Just visualizing the function values in each iterations.
